```python
from jinja2 import Template
```

```python
parameters=[
    {
        "table" : "catalog.schema.table",
        "alias" : "table",
        "cols"  : "table.col_name, table.col_name"
    },
    {
        "table" : "catalog.schema.table1",
        "alias" : "table1",
        "cols"  : "table1.col_name, table1.col_name",
        "condition" : "table.col_name= table1.col_name"
    },
    {
        "table" : "catalog.schema.table2",
        "alias" : "table2",
        "cols"  : "table2.col_name, table2.col_name",
        "condition" : "table.col_name= table2.col_name"
    }
]
```

```python
query_text= """


        SELECT
            {% for param in parameters %}
                {{ param.cols }}
                    {% if not loop.last %}
                        ,
                    {% endif %}
            {% endfor %}
        FROM
            {% for param in parameters %}
                {% if loop.first %}
                    {{ param['table'] }} AS {{ param['alias'] }}
                {% endif %}
            {% endfor %}
            {% for param in parameters %}
                {% if not loop.first %}
                    LEFT JOIN   
                        {{ param['table']}} AS {{ param['alias'] }}
                    ON
                        {{ param['condition'] }}
                {% endif %}
            {% endfor %}


"""
```

```python
jinja_sql_str=Template(query_text)
query=jinja_sql_str.render(parameters=parameters)
print(query)
```
